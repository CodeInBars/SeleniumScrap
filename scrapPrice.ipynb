```python
def pccomponentes(url):
    driver.get(url)
    time.sleep(30)
    containers = driver.find_element("xpath", '//*[@id="precio-main"]')
    return float(containers.text.replace("\n","").replace(",",".")[:-1])

def amazon(url):
    driver.get(url)
    time.sleep(30)
    containers = driver.find_elements("xpath",'//*[@id="corePrice_feature_div"]/div/span/span[2]/span[1]')
    print(len(containers))
    print(containers)
    if(len(containers) == 0):
        containers = driver.find_element("xpath",'//*[@id="corePriceDisplay_desktop_feature_div"]/div[1]/span/span[2]/span[1]')

    containers2 = driver.find_elements("xpath",'//*[@id="corePrice_feature_div"]/div/span/span[2]/span[2]')
    print(len(containers2))
    print(containers2)
    if(len(containers2) == 0):
        containers2 = driver.find_element("xpath",'//*[@id="corePriceDisplay_desktop_feature_div"]/div[1]/span/span[2]/span[2]')
        
    return float(containers.text+"."+containers2.text)

def rand_proxy():
    proxy = random.choice(proxyList.ips)
    return proxy

def recharge_driver():
    chrome_options = webdriver.ChromeOptions()
    proxy = rand_proxy()
    if(proxy!="None"):
        chrome_options.add_argument(f'--proxy-server={proxy}')
        return webdriver.Chrome(options=chrome_options)
    else:
        return webdriver.Chrome()
```

```python
listas = []
for urls in dic:
    flag1 = True
    flag2 = True
    price1 = 0
    price2 = 0
    trys = 0
    while(flag1):
        if(trys == 10):
            flag1 = False
            print("Problemas en el scrapeado del %s " % (urls[0]))
        try:
            driver = recharge_driver()
            price1 = pccomponentes(urls[1])
            flag1 = False
            driver.quit()
        except:
            trys += 1
            driver.quit()
    trys = 0
    while(flag2):
        print(trys)
        if(trys == 10):
            flag2 = False
            print("Problemas en el scrapeado del %s " % (urls[0]))
        try:
            driver = recharge_driver()
            price2 = amazon(urls[2])
            flag2 = False
            driver.quit()
        except Exception as e:
            trys += 1
            driver.quit()
    result = 0
    emoji = "❌"
    try:
        if(price1>price2):
            result = price1 - price2
            if(price2<=diccionario_prices.get(urls[0])):
                emoji = "✅"
        else:
            result = price2 - price1
            if(price1<=diccionario_prices.get(urls[0])):
                emoji = "✅"
    except:
        result = 0
        print("Fail in %s" % (urls))

    lista = [urls[0], price1, price2, result, diccionario_prices.get(urls[0]), emoji, today]
    listas.append(lista)
    time.sleep(120)

df = DataFrame(listas, columns=["Componente","Pccomponentes","Amazon", "Dif", "Lower", "To Build", "Date"])
print(df)
```
